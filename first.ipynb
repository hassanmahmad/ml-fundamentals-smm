```python
# import numpy as np

# Dimension of the problem
n = 10

# Create the vectors
x1 = np.linspace(0, 1,n)
# print(f'linspace: {x1}')
x2 = np.random.randn(n)
# print(f'randn: {x2}')

# Create the matrix
A = np.random.randn(n, n)
# print(f'randn A: {A}')

# Compute y1 and y2
y1 = A @ x1
y2 = A @ x2

# Compute y = A(x1 + x2)
y = A @ (x1 + x2)

# Check the equality
print(y)
print(y1 + y2)
```

```python
# Dimension of the problem
n = 3

# Create the vectors
x1 = np.linspace(0, 1,n)
# print(f'linspace: {x1}')
x2 = np.random.randn(n)

y1 = np.exp(x1)
y2 = np.exp(x2)

y = np.exp(x1+x2)

print(y)
print(y1*y2)

print(np.isclose(y,y1*y2))
```

```python
A = np.random.normal(3,3,(10_000_000,))
B = A[A>0]
B[B<0] = 0
print(B)

A = np.random.normal(3, 3, (10_000_000,))

# Create B with same shape as A
B = np.copy(A)  # or B = A.copy()
B[B < 0] = 0

print(B)
```

```python
# import numpy as np

# Set the dimension n
n = 4  # you can change this to any size you want

# Create matrix A with normally distributed values
A = np.random.normal(size=(n, n))

# Create matrix B using numpy's where function
# where(condition, value_if_true, value_if_false)
B = np.where(A > 0, A, 0)

print("Matrix A (normally distributed):")
print(A)
print("\nMatrix B (positive values only):")
print(B)
```
