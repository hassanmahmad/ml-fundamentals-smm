```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read data from a csv file
data = pd.read_csv('./US_births_2000-2014_SSA.csv')

# print(data.head())
total_date = []
i = 0
print(len(data))
while i < len(data):
    total_date.insert(i,i)
    i += 1
# print(data.head())
# print(total_date)
data['total_date'] = total_date
print(data.head())

# Visualize
plt.plot(data['births'], data['total_date'],'-',color='red')
plt.title('A plot of birth b/w 2000 to 2014')
plt.xlabel('Number of Births')
plt.ylabel('Date from Begining')
plt.grid()
plt.show()

plt.bar(data['day_of_week'],data['births'])
plt.xlabel('Day of Week')
plt.ylabel('# of Births')
plt.grid()
plt.show()

plt.bar(data['year'],data['births'])
plt.xlabel('Day of Week')
plt.ylabel('# of Births')
plt.grid()
plt.show()

plt.bar(data['total_date'],data['births'])
plt.xlabel('Day of Week')
plt.ylabel('# of Births')
plt.grid()
plt.show()
```
