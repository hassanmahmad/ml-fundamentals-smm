```python
import numpy as np
import matplotlib.pyplot as plt

# Setting up the dimension
n_max = 10

relative_errors = np.zeros((n_max, ))
two_norms = np.zeros((n_max, ))
inf_norms = np.zeros((n_max, ))

for n in range(1, n_max+1):
# Creating the matrix the vector
    A = np.random.randn(n,n) # n x n random matrix
    x_true = np.ones((n, ))   # n-dimensional vector of ones

    y = A @ x_true # Compute the term y s.t. x_true is a sol.

    # condition number with p=2
    cond_rand = np.linalg.cond(A, p=2)
    two_norms[n-1] = cond_rand
    # print(f"Condition number for A at p=2: {cond_rand}.")

    # condition number with p=inf
    cond_rand_inf = np.linalg.cond(A, p=np.inf)
    inf_norms[n-1] = cond_rand_inf
    # print(f"Condition number for A at p=inf: {cond_rand_inf}.")

    # Solve using np.linalg.solve()
    x_sol = np.linalg.solve(A, y)

    # Compute the relative error between the computed solution and the true solution
    error_rel = np.linalg.norm(x_sol - x_true, 2) / np.linalg.norm(x_true, 2)
    relative_errors[n-1] = error_rel
    # print(f"The relative error is {error_rel}")

# Visualize
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 2)
plt.plot(np.arange(1, n_max+1), two_norms)
plt.plot(np.arange(1, n_max+1), inf_norms)
plt.title('A plot of norm differenec')
plt.xlabel('n')
plt.ylabel('norm')
plt.legend(['2-Norm', 'Inf Norm'])
plt.grid()

plt.subplot(1, 2, 1)
plt.plot(np.arange(1, n_max+1), relative_errors)
plt.title('Relative Error over n')
plt.xlabel('n')
plt.ylabel('relative error')
plt.grid()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Setting up the dimension
n_arrange = np.arange(5, 35, 5)
# print(n_arrange)

i = 0
relative_errors = np.zeros((len(n_arrange), ))
two_norms = np.zeros((len(n_arrange), ))
inf_norms = np.zeros((len(n_arrange), ))

for n in n_arrange:
    
    # Creating the matrix the vector
    vander_vector = np.array(np.arange(1,n+1,1))
    A = np.vander(vander_vector, N=n) # n Vander matrix
    x_true = np.ones((n, ))   # n-dimensional vector of ones

    y = A @ x_true # Compute the term y s.t. x_true is a sol.

    # condition number with p=2
    cond_rand = np.linalg.cond(A, p=2)
    two_norms[i] = cond_rand
    # print(f"Condition number for A at p=2: {cond_rand}.")

    # condition number with p=inf
    cond_rand_inf = np.linalg.cond(A, p=np.inf)
    inf_norms[i] = cond_rand_inf
    # print(f"Condition number for A at p=inf: {cond_rand_inf}.")

    # Solve using np.linalg.solve()
    x_sol = np.linalg.solve(A, y)

    # Compute the relative error between the computed solution and the true solution
    error_rel = np.linalg.norm(x_sol - x_true, 2) / np.linalg.norm(x_true, 2)
    relative_errors[i] = error_rel
    i += 1
    # print(f"The relative error is {error_rel}")

# Visualize
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 2)
plt.plot(n_arrange, two_norms)
plt.plot(n_arrange, inf_norms)
plt.title('A plot of norm differenec with Vander Matrix')
plt.xlabel('n')
plt.ylabel('norm')
plt.legend(['2-Norm', 'Inf Norm'])
plt.grid()

plt.subplot(1, 2, 1)
plt.plot(n_arrange, relative_errors)
plt.title('Relative Error over n')
plt.xlabel('n')
plt.ylabel('relative error')
plt.grid()

plt.show()
```
